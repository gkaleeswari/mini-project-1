```python
from itertools import combinations

def find_subsets_with_sum_zero(nums):
    subsets = []

    for combo in combinations(nums, 5):
        if sum(combo) == 0:
            subsets.append(combo)

    return subsets

nums = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

result = find_subsets_with_sum_zero(nums)
for subset in result:
    print(subset)
```

```python
from itertools import combinations

def find_subsets_with_sum(nums):
    subsets = []
    for size in range(3, 7):

        for combo in combinations(nums, size):
            if sum(combo) == 0:
                subsets.append(combo)

    return subsets

nums = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

result = find_subsets_with_sum(nums)
for subset in result:
    print(subset)
```
